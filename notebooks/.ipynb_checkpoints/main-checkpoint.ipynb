#### Synthetic Data Generation

```python
#Step 1: Generate Synthetic Data and save as df
def get_rdata(x_botleft, y_botleft, x_topright, y_topright, pp_square):
    points = np.column_stack((
    np.random.uniform(x_botleft, x_topright, pp_square),
    np.random.uniform(y_botleft, y_topright, pp_square),    
    ))
    return points

c0 = np.vstack((
    get_rdata(0,0,1,1,400),
    get_rdata(0,1,1,2,400),
    get_rdata(1,1,2,2,400)
))
#print(c0_1[:,1]) #only y
#print(c0_1[:,0]) #only x
c1 = np.vstack((
    get_rdata(1,0,2,1,400),
    get_rdata(2,0,3,1,400),
    get_rdata(2,1,3,2,400)
))
```

```python
#Step 2: Plot synthetic data
#plt.scatter(c0[:,0],c0[:,1],color="blue",marker="x", label="class 0")
#plt.scatter(c1[:,0],c1[:,1],color="green",marker="x", label="class 1")
#plt.grid(True)
#plt.title("binary class data")
#plt.legend(loc="upper right")
#plt.show()
```

```python
#step 3: lets add the true decision boundary for fun!
db = np.vstack(((1,0),(1,1),(2,1),(2,2)))
```

```python
#step 4: plot again
plt.scatter(c0[:,0],c0[:,1],color="blue",marker="x", label="class 0")
plt.scatter(c1[:,0],c1[:,1],color="green",marker="x", label="class 1")
#plt.plot(db[:,0],db[:,1],color="red",linestyle="-", label="db")
plt.grid(True)
plt.title("binary class data")
plt.legend(loc="upper right")
plt.show()
```

```python
#step 4: create dataframes
df_c0 = pd.DataFrame(c0, columns=["x0", "x1"])
df_c0["class"] = 0
df_c1 = pd.DataFrame(c1, columns=["x0", "x1"])
df_c1["class"] = 1
df = pd.concat([df_c0, df_c1], ignore_index=True)
df = df.sample(frac=1).reset_index(drop=True)
```

```python
X = df.drop(columns=["class"])
y = df["class"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=67, stratify=y)

X_train.info()
```

#### Model Architecture

```python
model = Sequential([
    Input(shape=(X_train.shape[1],)),
    Dense(32, activation="relu"),
    Dense(32, activation="relu"),
    Dense(32, activation="relu"),
    Dense(32, activation="relu"),
    Dense(1, activation="sigmoid")
])

model.compile(optimizer=Adam(learning_rate=0.001), loss="binary_crossentropy", metrics=["accuracy"])
model.fit(X_train, y_train, epochs=100, batch_size=32, verbose=1)
y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5).astype(int)
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

#### Testing a single value

#### Building decision boundary plot

```python
#step 1: get dimensions of features pace, adding a buffer of 1 either side for nicer visualisation
x0_min, x0_max = (X.iloc[:,0].min() - 1, X.iloc[:,0].max() + 1)
x1_min, x1_max = (X.iloc[:,1].min() - 1, X.iloc[:,1].max() + 1)
#step 2: Create list of x points and y points corresponding to each over via meshgrid
xx0, xx1 = np.meshgrid(np.linspace(x0_min, x0_max, 100), np.linspace(x1_min, x1_max, 100))
# e.g. x0 = [0,1,2], x1 = [10,20,30] -> xx0 = [[0,1,2],[0,1,2],[0,1,2]], xx1 = [[10,10,10],[20,20,20],[30,30,30]]
#step 3: flatten into 1d arrays using .ravel() and then concatenate (match up x0 to x1 points) via c_
grid = np.c_[xx0.ravel(), xx1.ravel()]
grid_df = pd.DataFrame(grid, columns=X.columns)
predictions = model.predict(grid_df, verbose=0)
print(predictions)
predictions = (predictions > 0.5).astype(int).reshape(xx0.shape)

plt.contourf(xx0, xx1, predictions, alpha=0.3, cmap="coolwarm")
plt.scatter(X_train.iloc[:,0], X_train.iloc[:,1], c=y_train, cmap="coolwarm", edgecolor='k',alpha=0.1)
plt.show()


```
