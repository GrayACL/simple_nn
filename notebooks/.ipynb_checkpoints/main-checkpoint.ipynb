```python
print("new project!")
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Input
from sklearn.metrics import accuracy_score
# --- TODO ---
#get validation set
#plot output
```

```python
#Step 1: Generate Synthetic Data and save as df
def get_rdata(x_botleft, y_botleft, x_topright, y_topright, pp_square):
    points = np.column_stack((
    np.random.uniform(x_botleft, x_topright, pp_square),
    np.random.uniform(y_botleft, y_topright, pp_square),    
    ))
    return points

c0 = np.vstack((
    get_rdata(0,0,1,1,100),
    get_rdata(0,1,1,2,100),
    get_rdata(1,1,2,2,100)
))
#print(c0_1[:,1]) #only y
#print(c0_1[:,0]) #only x
c1 = np.vstack((
    get_rdata(1,0,2,1,100),
    get_rdata(2,0,3,1,100),
    get_rdata(2,1,3,2,100)
))
```

```python
#step 4: plot again
plt.scatter(c0[:,0],c0[:,1],color="blue",marker="x", label="class 0")
plt.scatter(c1[:,0],c1[:,1],color="green",marker="x", label="class 1")
#plt.plot(db[:,0],db[:,1],color="red",linestyle="-", label="db")
plt.grid(True)
plt.title("binary class data")
plt.legend(loc="upper right")
plt.show()
```

```python
#step 4: create dataframes
df_c0 = pd.DataFrame(c0, columns=["X", "Y"])
df_c0["class"] = 0
df_c1 = pd.DataFrame(c1, columns=["X", "Y"])
df_c1["class"] = 1
df = pd.concat([df_c0, df_c1], ignore_index=True)
df = df.sample(frac=1).reset_index(drop=True)
```

```python
X = df.drop(columns=["class"])
y = df["class"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=67, stratify=y)

X_train.info()
```

```python
model = Sequential([
    Input(shape=(X_train.shape[1],)),
    Dense(16, activation="relu"),
    Dense(1, activation="sigmoid")
])
```

```python
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
model.fit(X_train, y_train, epochs=10, batch_size=8)
y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5).astype(int)
accuracy = accuracy_score(y_test, y_pred)
accuracy
```
